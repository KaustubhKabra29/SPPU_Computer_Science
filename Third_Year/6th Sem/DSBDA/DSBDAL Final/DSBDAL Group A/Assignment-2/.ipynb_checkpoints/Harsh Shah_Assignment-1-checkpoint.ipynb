# Detecting and Removing Outliers using Z-Score Method

```python
def detect(data):
    outliers=[]
    threshhold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        zscore=(i-mean)/std
        if(np.abs(zscore)>3):
            outliers.append(i)
    return outliers
```

```python
YearsAtCompany=detect(data['YearsAtCompany'])
print(YearsAtCompany)
```

```python
TotalWorkingYears=detect(data['TotalWorkingYears'])
print(TotalWorkingYears)
```

```python
def replace(data):
    outliers=[]
    threshhold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        zscore=(i-mean)/std
        if(zscore>3):
            data.replace(i,3,inplace=True)
            outliers.append(i)
        if(zscore<-3):
            data.replace(i,-3,inplace=True)
            outliers.append(i)
    return outliers
```

```python
replace(data['YearsAtCompany'])
```

```python
replace(data['TotalWorkingYears'])
```

```python
'''
As we can observe, there are a few outliers even after removal in z-score
This is because we replace the outliers with 3 or -3
As a result, the standard deviation and mean changes which further changes the z-score
Thus, some outliers are still present

'''
sns.boxplot(data['YearsAtCompany'])
```

```python
sns.boxplot(data['TotalWorkingYears'])
```

```python
'''
The solution is to run the z-score algorithm for a few number of times
This reduces the number of outliers further more.

'''
for i in range (0,5):
    YearsAtCompany=detect(data['YearsAtCompany'])
    replace(data['YearsAtCompany'])
```

```python
'''
Now we can see we have very less outliers

'''
sns.boxplot(data['YearsAtCompany'])
```

# Detecting and Removing Outliers using IRQ Method

```python
def detect_remove_IQR(data_values):
    outliers=[]
    q1=data_values.quantile(0.25)
    q3=data_values.quantile(0.75)
    IQR=q3-q1
    lower_bound=q1-(1.5*IQR)
    upper_bound=q3+(1.5*IQR)
    for i in data_values:
        if(i<lower_bound or i>upper_bound):
            outliers.append(i)
    df=pd.DataFrame(data[(data_values<upper_bound) & (data_values>lower_bound)])
    return df
```

```python
data=detect_remove_IQR(data['YearsSinceLastPromotion'])
data.head()
```

```python
sns.boxplot(data['YearsSinceLastPromotion'])
```
