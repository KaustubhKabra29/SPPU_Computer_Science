```python
def detect(data):
    outliers=[]
    threshhold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        zscore=(i-mean)/std
        if(np.abs(zscore)>3):
            outliers.append(i)
    return outliers
```

```python
absences=detect(copy_data['absences'])
print(absences)
```

```python
def replace(data):
    outliers=[]
    threshhold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        zscore=(i-mean)/std
        if(zscore>3):
            data.replace(i,3,inplace=True)
            outliers.append(i)
        if(zscore<-3):
            data.replace(i,-3,inplace=True)
            outliers.append(i)
    return outliers
```

```python
replace(copy_data['absences'])
```
