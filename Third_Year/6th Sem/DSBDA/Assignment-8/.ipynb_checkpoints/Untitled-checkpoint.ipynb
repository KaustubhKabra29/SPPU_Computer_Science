```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data
```

```python
plt.figure(figsize=(16,10))
sns.histplot(data=data['Fare'])
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.boxplot(y=data['Age'], x=data['Sex'], hue=data['Survived'])
plt.plot()
```

1) Number of females survived is more than female casualties.<br>
2) Number of males survived is less than the male casualties.<br>
3) Hence, first priority was to save females then males.<br>
